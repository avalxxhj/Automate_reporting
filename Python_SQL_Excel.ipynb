```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
```

```python
file = 'Employee-Attribution.xlsx'
output = 'output.xlsx'
```

```python
engine = create_engine('sqlite://', echo=False)
df = pd.read_excel(file, sheet_name='Emplyee_Data_Set')
df.to_sql('employees', engine, if_exists='replace', index=False)

results = engine.execute("Select * from employees where EducationField in('Life Sciences', 'Medical') \
                            And DistanceFromHomoe < 25 \
                            And DailyRate > 300 \
                            ")

final=pd.DataFrame(results, columns=df.columns)
final.to_excel(output, index=False)
final
```
